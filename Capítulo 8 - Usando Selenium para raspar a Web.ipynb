## Introdução ao Selenium
Como mencionei, o Selenium é um framework de aplicação web que pode ser usado para atividades de web scraping. Também pode ser usado como uma ferramenta de automação do navegador.

A automação de tarefas ou atividades relacionadas a aplicativos da Web, como as da lista a seguir, envolve a execução dessas tarefas sem o envolvimento direto de seres humanos:
* Navegando
* Clicar em links
* Salvando capturas de tela
* Baixando imagens
* Preenchendo modelos HTML \<form> e muitas outras atividades

O Selenium fornece um navegador web como interface ou ferramenta automatizada. Com a automatização da ação de navegação, o Selenium também pode ser utilizado em web scraping. Serviços da Web dinâmicos ou seguros que usam JavaScript, cookies, scripts e assim por diante são carregados, testados e até rastreados e raspados com a ajuda do Selenium.

O Selenium é de código aberto e pode ser acessado em várias plataformas. Vários navegadores da Web podem ser usados para testes usando bibliotecas disponíveis para linguagens de programação como Java e Python. As bibliotecas são usadas para criar scripts que interagem com o Selenium para realizar a automação baseada em navegador.

Embora o uso do Selenium no teste de aplicativos tenha muitas vantagens quando se trata de ações como rastreamento e raspagem, ele também tem suas desvantagens, como **tempo e consumo de memória**. O Selenium é extensível e eficaz, mas é lento na execução das suas ações e consome muito de espaço de memória.

### Projetos de Selenium

O Selenium consiste em vários componentes ou ferramentas que também são conhecidos como projetos Selenium, o que a torna uma estrutura completa para testes de aplicativos baseados na web. Veremos agora alguns dos principais componentes desses projetos do Selenium.

#### Selenium Webdriver
Selenium WebDriver é um componente do Selenium que é usado para automatizar o navegador. A automatização do navegador pode ser realizada fornecendo comandos com várias ligações de linguagem disponíveis para Java, Python, JavaScript e assim por diante, usando drivers de terceiros, como driver do Google Chrome, Mozilla Driver Gecko e Opera (`https://github.com/mozilla/geckodriver/`). O Selenium WebDriver não tem dependência externa de nenhum outro software ou servidor.

O WebDriver é uma API orientada a objetos com recursos atualizados que superam e abordam as limitações das versões anteriores do Selenium e do **Selenium Remote Control (RC)**.

#### Selenium RC
O Selenium RC é um servidor programado em Java. Ele usa HTTP para aceitar comandos para o navegador e é usado para testar aplicativos da Web complexos baseados em AJAX.

O Selenium RC foi oficialmente obsoleto após o lançamento do Selenium 2 (Selenium versão 2). No entanto, o WebDriver contém os principais recursos do Selenium RC.

#### Selenium Grid
O Selenium Grid também é um servidor que permite que testes sejam executados paralelamente em várias máquinas, em vários navegadores e sistemas operacionais, distribuindo a carga do sistema e reduzindo problemas de desempenho, como consumo de tempo.

Testes complexos foram usados para processar o Selenium RC e o Selenium Grid juntos. Desde o lançamento da versão 2.0, o servidor Selenium agora tem suporte interno para WebDriver, Selenium RC e Selenium Grid.

#### Selenium IDE

Um ambiente de desenvolvimento integrado (IDE) Selenium de código aberto é usado para criar casos de teste com o Selenium. É basicamente uma extensão de navegador da web disponível com recursos como a capacidade de gravar e reproduzir a automação da web por meio de uma interface gráfica do usuário (GUI).

A seguir estão alguns dos principais recursos do Selenium IDE:
* Extensível e fácil de usar para depuração;
* Testes resilientes;
* Suporte entre navegadores; e
* Pode criar scripts que podem executar comandos e suportar estruturas de fluxo de controle.

### Explorando o Selenium
Nesta seção, usaremos e apresentaremos várias propriedades para `webdriver` e `webdriver.Chrome`, enquanto analisamos alguns casos reais. As seções a seguir ilustrarão o uso do Selenium e explorarão as suas principais propriedades.

#### Acessando as propriedades do navegador
Para começar, vamos importar o `webdriver` do selenium e definir um caminho para `chromedriver.exe` — vamos chamá-lo de `chromedriver_path`. O caminho criado será necessário para carregar o Google Chrome. Dependendo da localização do aplicativo, o caminho completo para `chromedriver.exe` deve ser mencionado e é obrigatório para uma implementação bem-sucedida:

```python
from selenium import webdriver
import re

chromedriver_path = 'chromedriver.exe'

# Provendo acesso ao driver

driver = webdriver.Chrome(executable_path=chromedriver_path)

# Acessando uma página utilizando o driver

url = 'https://www.python.org'
driver.get(url)

# Uma vez que temos acesso à pagina pelo driver podemos acessar as propriedades.

print(f"Título: {driver.title}")
print(f"\nURL da Página atual: {driver.current_url}")
```

Conforme visto no código anterior, o título da página está disponível usando `driver.title` e a URL da página atual é encontrada com `driver.current_url`.
 A frase `current_url` pode ser usada para verificar se algum redirecionamento de URL ocorreu após o carregamento do URL inicial. Vamos salvar uma captura de tela da página com uma condição que é verificada usando `search()` da biblioteca Python, `re`:

```python
#check if pattern matches the current url loaded
if re.search(r'python.org',driver.current_url):
	driver.save_screenshot("pythonorg.png") #save screenshot with provided name
	print("Python Screenshot Saved!")
```

```python
# Para explorar mais, vamos coletar os cookies da web de https://www.python.org.
# A frase get_cookies() é usada para recuperar cookies, da seguinte forma:

#Capturando as informações de cookies
cookies = driver.get_cookies()
print("Cookies obtained from python.org")
print(cookies)
```

```python
# A página fonte pode ser obtida usando <driver.page_source>

print(driver.page_source)
```

```python
# A página pode ser recarregada usando <driver.refresh()>
driver.refresh()
```

Com os recursos que foram acessados usando o driver no código anterior, vamos continuar carregando, fazendo capturas de tela e acessando cookies de `https://www.google.com` usando o seguinte código:

```python
url = 'https://www.google.com'
driver.get(url)
print(f"Título: {driver.title}")
print(f"URL da Página Atual: {driver.current_url}")

if re.search(r'google.com', driver.current_url):
	driver.save_screenshot('google.png')
	print('Tela do Google salva!')

cookies = driver.get_cookies()
```

A ação realizada com `http://google.com` ocorrerá na mesma janela do navegador que foi usada para acessar `http://python.org`. Com isso, já podemos realizar ações usando o histórico do navegador (ou seja, usaremos os botões Voltar e Avançar que estão disponíveis no navegador) e recuperar a URL, conforme mostrado no código a seguir:

```python
print("Current Page URL: ",driver.current_url)
driver.back() #History back action
print("Page URL (Back): ",driver.current_url)
driver.forward() #History forward action
print("Page URL (Forward): ",driver.current_url)
```

Após a execução bem-sucedida do código, é recomendável fechar e sair do driver para liberar recursos do sistema. Podemos executar as ações de encerramento usando as seguintes funções:

```python
driver.close() #close browser
driver.quit() #quit webdriver
```

```python
# Execução completa das ações
from selenium import webdriver
import re
chrome_path='chromedriver'
driver = webdriver.Chrome(executable_path=chrome_path) #print(type(driver))
driver.get('https://www.python.org')
print("Title: ",driver.title)
print("Current Page URL: ",driver.current_url)
if re.search(r'python.org',driver.current_url):
    driver.save_screenshot("pythonorg.png")
print("Python Screenshot Saved!")
cookies = driver.get_cookies()
print(driver.page_source)
driver.refresh()
driver.get('https://www.google.com')
print("Title: ",driver.title)
print("Current Page URL: ",driver.current_url)
if re.search(r'google.com',driver.current_url):
    driver.save_screenshot("google.png")
print("Google Screenshot Saved!")
cookies = driver.get_cookies()
print("Current Page URL: ",driver.current_url)
driver.back()
print("Page URL (Back): ",driver.current_url)
driver.forward()
print("Page URL (Forward): ",driver.current_url)
driver.close()
driver.quit()
```

Com o número de itens e a contagem de produtos encontrados, isso transmite uma mensagem de sucesso ao nosso processo de busca. Agora, podemos continuar procurando produtos usando XPath, seletores CSS e assim por diante:

#### Exemplo 2 – raspagem de informações do livro

Neste exemplo iremos capturar as seguintes informações:
* titulo
* titulo completo
* preço
* stock
* imagem
* rating
* url

```python
# Vamos capturar uma seção específica <Food and Drink>
# Depois vamos clicar

from selenium import webdriver
chrome_path='chromedriver'
driver = webdriver.Chrome(executable_path=chrome_path)
driver.get('http://books.toscrape.com/index.html')
driver.find_element_by_link_text('Food and Drink').click()
print("Current Page URL: ", driver.current_url)
totalBooks = driver.find_element_by_xpath("//*[@id='default']//form/strong[1]")
print("Found: ", totalBooks.text)
```

Para lidar com várias páginas encontradas durante a iteração, `NoSuchElementException` de `selenium.common.exceptions` será importado:
`from selenium.common.exceptions import NoSuchElementException`

Como usaremos o botão de paginação a seguir, `NoSuchElementException` será útil para lidar com a condição se não forem encontradas mais próximas ou páginas. Como visto no código a seguir, a próxima opção de paginação está localizada na página e processada com a ação `click()`. Esta ação carregará o URL contém para o navegador, e a iteração continuará até que *next* não seja localizado ou encontrado na página, capturado pelo bloco `except` no código:

```python
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException

chrome_path = 'chromedriver'
driver = webdriver.Chrome(executable_path=chrome_path)
driver.get('http://books.toscrape.com/index.html')

dataSet = []
# select: Food and Drink
driver.find_element_by_link_text("Food and Drink").click()
print("Current Page URL: ", driver.current_url)
totalBooks = driver.find_element_by_xpath("//*[@id='default']//form/strong[1]")
print("Found: ", totalBooks.text)

page = True
while page:
    listings = driver.find_elements_by_xpath("//*[@id='default']//ol/li[position()>0]")
    for listing in listings:
        url = listing.find_element_by_xpath(".//article[contains(@class,'product_pod')]/h3/a").get_attribute('href')
        title = listing.find_element_by_xpath(".//article[contains(@class,'product_pod')]/h3/a").text
        titleLarge = listing.find_element_by_xpath(".//article[contains(@class,'product_pod')]/h3/a").get_attribute(
            'title')
        price = listing.find_element_by_xpath(".//article/div[2]/p[contains(@class,'price_color')]").text
        stock = listing.find_element_by_xpath(".//article/div[2]/p[2][contains(@class,'availability')]").text
        image = listing.find_element_by_xpath(
            ".//article/div[1][contains(@class,'image_container')]/a/img").get_attribute('src')
        starRating = listing.find_element_by_xpath(".//article/p[contains(@class,'star-rating')]").get_attribute(
            'class')
        dataSet.append([titleLarge, title, price, stock, image, starRating.replace('star-rating ', ''), url])

    try:
        #Check for Pagination with text 'next'
        driver.find_element_by_link_text('next').click()
        continue
    except NoSuchElementException:
        page = False

print("Completed")

print(dataSet)

driver.close()
driver.quit()
```
