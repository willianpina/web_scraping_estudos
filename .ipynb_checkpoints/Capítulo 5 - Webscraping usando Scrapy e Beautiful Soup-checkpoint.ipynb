## Web Scraping using Beautiful Soup

```python
import bs4
bs4.__version__
```

```python
dir(bs4.BeautifulSoup)
```

```python
from bs4 import SoupStrainer, BeautifulSoup
from urllib.request import urlopen
```

```python
tagsA = SoupStrainer('a')
content = open('pythonpage.html').read()
soupA = BeautifulSoup(content, 'html', parse_only=tagsA)
print(type(soupA))
```

```python
print(soupA)
```

```python
print(soupA.prettify())
```

## Pesquisando, percorrendo e iterando

```python
# Procura um elemento ou tag <a> na primeira ocorrência.
print(soupA.find('a'))

# Procura um elemento <a> com o atributo e valor
print(soupA.find(name='a'))

print(soupA.find('a', attrs={'class':'jump-link'}))

# Procura o <a> com a classe e a filha e texto chamado 'network'
print(soupA.find('a',attrs={'class':'jump-link'}, text="network"))
```

```python
# Find_all() procura todas as ocorrências da tag

print(soupA.find_all('a'))
```

```python
# Limitando o número de ocorrências
print(soupA.find_all('a', limit=2)) # Retorna somente 2 ocorrências.
```

```python
for a in soupA.find_all('a', limit=2):
    print(a)
```
